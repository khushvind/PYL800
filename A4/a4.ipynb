```python
import matplotlib.pyplot as plt
```

## Euler's Method

```python
def euler_method(f, y0, t0, tn, h):
    """
    Euler's method for solving a first-order ordinary differential equation (ODE).

    Parameters:
    - f: Function representing the ODE dy/dt = f(t, y)
    - y0: Initial value of y at t = t0
    - t0: Initial time
    - tn: Final time
    - h: Step size

    Returns:
    - t_values: List of time values
    - y_values: List of corresponding y values
    """

    t_values = [t0]
    y_values = [y0]

    t = t0
    y = y0

    while t < tn:
        y += h * f(t, y)
        t += h

        t_values.append(t)
        y_values.append(y)

    return t_values, y_values

# Example usage:
# Suppose you have an ODE dy/dt = t + y with initial condition y(0) = 1
def example_ode(t, y):
    return 2*t+-1

# Solve the ODE using Euler's method
t_values, y_values = euler_method(example_ode, y0=0, t0=0, tn=1, h=0.1)

# Plot the results
plt.plot(t_values, y_values, label='Euler\'s Method')
plt.xlabel('Time (t)')
plt.ylabel('y(t)')
plt.title('Euler\'s Method for ODE')
plt.legend()
plt.show()
```

## RK2 Method

```python
def rk2_method(f, y0, t0, tn, h):
    """
    Runge-Kutta method of order 2 (RK2) for solving a first-order ODE.

    Parameters:
    - f: Function representing the ODE dy/dt = f(t, y)
    - y0: Initial value of y at t = t0
    - t0: Initial time
    - tn: Final time
    - h: Step size

    Returns:
    - t_values: List of time values
    - y_values: List of corresponding y values
    """

    t_values = [t0]
    y_values = [y0]

    t = t0
    y = y0

    while t < tn:
        k1 = h * f(t, y)
        k2 = h * f(t + h/2, y + k1/2)

        y += k2
        t += h

        t_values.append(t)
        y_values.append(y)

    return t_values, y_values

# Example usage:
# Suppose you have an ODE dy/dt = t + y with initial condition y(0) = 1
def example_ode(t, y):
    return 2*t+-1

# Solve the ODE using RK2
t_values_rk2, y_values_rk2 = rk2_method(example_ode, y0=0, t0=0, tn=1, h=0.1)

# Plot the results
plt.plot(t_values_rk2, y_values_rk2, label='RK2 Method')
plt.xlabel('Time (t)')
plt.ylabel('y(t)')
plt.title('Runge-Kutta Method of Order 2 (RK2) for ODE')
plt.legend()
plt.show()
```

# RK3

```python
def rk4_method(f, y0, t0, tn, h):
    """
    Runge-Kutta method of order 4 (RK4) for solving a first-order ODE.

    Parameters:
    - f: Function representing the ODE dy/dt = f(t, y)
    - y0: Initial value of y at t = t0
    - t0: Initial time
    - tn: Final time
    - h: Step size

    Returns:
    - t_values: List of time values
    - y_values: List of corresponding y values
    """

    t_values = [t0]
    y_values = [y0]

    t = t0
    y = y0

    while t < tn:
        k1 = h * f(t, y)
        k2 = h * f(t + h/2, y + k1/2)
        k3 = h * f(t + h/2, y + k2/2)
        k4 = h * f(t + h, y + k3)

        y += (k1 + 2*k2 + 2*k3 + k4) / 6
        t += h

        t_values.append(t)
        y_values.append(y)

    return t_values, y_values

# Example usage:
# Suppose you have an ODE dy/dt = t + y with initial condition y(0) = 1
def example_ode(t, y):
    return 2*t-1

# Solve the ODE using RK4
t_values_rk4, y_values_rk4 = rk4_method(example_ode, y0=1, t0=0, tn=1, h=0.1)

# Plot the results
plt.plot(t_values_rk4, y_values_rk4, label='RK4 Method')
plt.xlabel('Time (t)')
plt.ylabel('y(t)')
plt.title('Runge-Kutta Method of Order 4 (RK4) for ODE')
plt.legend()
plt.show()
```
